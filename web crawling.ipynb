```python
from lxml import html
import requests
```

```python
page = requests.get('http://www.allabolag.se/5567164800/incontext-ab')
page.content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
list(soup.children)
```

```python
company_table = soup.find('table', class_="company-table") 
data_rows = company_table.find_all('tr')

for row in data_rows:
    if row.th and row.th.text:
        name = row.th.text.strip()
        values = row.find_all('td')
        print(name)
        for value in values:
            clean_value = value.text.strip()
            print(clean_value)

```
